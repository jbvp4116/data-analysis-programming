## 배열 결합 연습

```python
a = np.arange(1, 4)
b = np.arange(4, 7)
c = np.arange(7, 10)
a, b, c
```

```python
p = np.arange(1, 7).reshape(2, 3)
q = np.arange(11, 17).reshape(2, 3)
r = np.arange(21, 27).reshape(2, 3)
p, q, r
```

```python
np.vstack((a, b, c)) # 1D -> 2D
```

```python
np.vstack((p, q, r)) #2D 유지
```

```python
np.hstack((a, b, c))
```

```python
np.hstack((p, q, r))
```

```python
a, b, c
```

```python
np.column_stack((a, b, c)) # 1D -> 2D
```

```python
np.column_stack((p, q, r)) # 2D 유지
```

```python
np.concatenate((a, b, c))
```

```python
np.concatenate((p, q, r))
```

```python
np.concatenate((p, q, r), axis = 1)
```

```python
np.concatenate((p, q, r), axis = None)
```

```python
np.c_[a, b, c]
```

```python
np.c_[p, q, r]
```

```python
np.r_[a, b, c]
```

```python
np.r_[p, q, r]
```

```python
np.concatenate((a, b, c))
```
